<h2 style="color:purple"><b>fandango's movie rating review analysis </b><h2>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fandango=pd.read_csv('fandango_scrape.csv')
```

```python
#explore the data
```

```python
fandango.head()
```

```python
fandango.dtypes
```

```python
#Relation between movie rating and votes
```

```python
plt.figure(figsize=(14,6),dpi=900)
sns.scatterplot(x='RATING',y='VOTES',data=fandango,marker='o',s=100)
plt.title('Relation between rating and votes')
plt.show()
```

```python
fandango['YEAR']=fandango['FILM'].apply(lambda title:title.split('(')[-1].replace(')',''))
```

```python
fandango['YEAR']=fandango['YEAR'].str[:4]
fandango['YEAR'].head(5)
```

```python
fandango
```

```python
#Count of movies per year
```

```python
sns.countplot(data=fandango,x='YEAR')
```

```python
fandango.YEAR.value_counts()
```

```python
#count of movies having zero votes
len(fandango[fandango['VOTES']==0])
```

```python
#creating data frames containing more than zero votes
fan_voted=fandango[fandango.VOTES>0]
len(fan_voted)
```

```python
fan_voted
```

```python
#comparison between actual rating and stars displayed
plt.figure(figsize=(14,6),dpi=1000)
sns.kdeplot(data=fan_voted,x='RATING',clip=[0,5],fill=True,label='True Rating')
sns.kdeplot(data=fan_voted,x='STARS',clip=[0,5],fill=True,label='Stars displayed')
plt.legend(loc=(1.05,0.5))
```

```python
fan_voted['STARS_DIFF']=(fan_voted['STARS']-fan_voted['RATING']).round(2)
fan_voted
```

```python
#plot showing the number of times difference between actual rating and stars displayed
```

```python
plt.figure(figsize=(15,6),dpi=1000)
sns.countplot(data=fan_voted,x='STARS_DIFF')
```

```python
fan_voted[fan_voted.STARS_DIFF==1]
```

<h3 style='color:green'><b>conclusion :only one movie has a stars displayed one greater than actual rating

<h2 style='color:purple'>comparison of fandango's rating with other sites

```python
all_sites=pd.read_csv('fandango_score_comparison.csv')
all_sites.head(10)
```

```python
all_sites.info()
```

```python
all_sites.describe()
```

```python
#analysis of RottenTomatoes critics vs user review
```

```python
plt.figure(figsize=(14,6),dpi=1000)
sns.scatterplot(data=all_sites,x='RottenTomatoes',y='RottenTomatoes_User',s=100)
```

```python
#calculatiing the difference between rottentomatoes score by critics and user
```

```python
all_sites['Rotten-diff']=all_sites['RottenTomatoes']-all_sites['RottenTomatoes_User']
all_sites['Rotten-diff'].apply(abs).mean()
```

```python
#distribution showing the difference between rottentomatoes critics and user score
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(data=all_sites,x='Rotten-diff',kde=True,bins=25)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(x=all_sites['Rotten-diff'].apply(abs),kde=True,bins=25)
```

```python
#Top 10 Movies with critics score higher than users on average
all_sites.nlargest(10,'Rotten-diff')['FILM']
```

```python
#Top 10 Movies with critics score lower than users on average
all_sites.nsmallest(10,'Rotten-diff')['FILM']
```

```python
#analysis of Metacritic critics vs Metacritic user review
```

```python
plt.figure(figsize=(14,6),dpi=150)
sns.scatterplot(data=all_sites,x='Metacritic',y='Metacritic_User',s=100)
```

```python
#conclusion :metacritic critics and user rating is almost same
```

```python
#Relationship between vote counts of metacritic user and IMDB
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.scatterplot(data=all_sites,x='Metacritic_user_vote_count',y='IMDB_user_vote_count')
```

```python
#sites with highest IMDB user vote count
all_sites.nlargest(1,'IMDB_user_vote_count')
```

```python
#sites with highest metacritic user vote count
all_sites.nlargest(1,'Metacritic_user_vote_count')
```

```python
#Fandango Score vs All sites score analysis
```

```python
#combining fandango movie rating with all sites
df=pd.merge(fandango,all_sites,on='FILM',how='inner')
```

```python
df
```

```python
df.describe().transpose()['max']
```

```python
#Normalisisng all sites rating 
```

```python
df['RT_Norm']=np.round(df['RottenTomatoes']/20,1)
df['RT_user_Norm']=np.round(df['RottenTomatoes_User']/20,1)
df['Metacritic_Norm']=np.round(df['Metacritic']/20,1)
df['Metacritic_User_Norm']=np.round(df['Metacritic_User']/2,1)
df['IMDB_Norm']=np.round(df['IMDB']/2,1)
```

```python
df.head()
```

```python
df.columns
```

```python
norm_scores=df[['STARS', 'RATING','RT_Norm', 'RT_user_Norm','Metacritic_Norm','Metacritic_User_Norm', 'IMDB_Norm']]
```

```python
norm_scores.head()
```

```python
#We already know that stars being displayed is higher than actual rating
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.kdeplot(data=norm_scores,shade=True)
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.histplot(norm_scores,bins=20)
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.kdeplot(data=norm_scores[['RT_Norm','STARS']],clip=[0,5],shade=True)
```

<h2 style='color:green'>Conclusion :From the plot we can conclude that rating beings displayed in fandango websites are way higher than the other websites rating.RottenTomatoes are rating the movie uniformly
    

```python
#worst movie rated across all sites
```
